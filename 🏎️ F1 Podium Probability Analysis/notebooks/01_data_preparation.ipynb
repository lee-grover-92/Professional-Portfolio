## 1. Setup  
Import core libraries and configure display settings for cleaner output.

```python
import pandas as pd
import numpy as np
import os

pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
```

## 2. Load Raw Data  
Read key CSV files into pandas DataFrames. These datasets cover drivers, constructors, races, results, qualifying sessions, and driver profiles.

```python
DATA_PATH = '../data/raw/'

# Load key datasets
drivers = pd.read_csv(os.path.join(DATA_PATH, 'drivers.csv'))
constructors = pd.read_csv(os.path.join(DATA_PATH, 'constructors.csv'))
races = pd.read_csv(os.path.join(DATA_PATH, 'races.csv'))
results = pd.read_csv(os.path.join(DATA_PATH, 'results.csv'))
qualifying = pd.read_csv(os.path.join(DATA_PATH, 'qualifying.csv'))
F1Drivers = pd.read_csv(os.path.join(DATA_PATH, 'F1Drivers_Dataset.csv'))
```

## 3. Initial Inspection  
Preview the first few rows of each dataset to understand structure and key fields.

```python
# Display first few rows of each dataset for inspection
print("Drivers:")
display(drivers.head())

print("Constructors:")
display(constructors.head())

print("Races:")
display(races.head())

print("Results:")
display(results.head())

print("Qualifying:")
display(qualifying.head())

print("F1Drivers Dataset:")
display(F1Drivers.head())
```

## 4. Data Cleaning  
This section performs basic cleaning steps to prepare the datasets for merging.  
Cleaning includes:
- handling missing values  
- fixing data types  
- removing duplicates  
- standardising naming conventions  
- light normalisation of key fields  

```python
def clean_column_names(df):
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(' ', '_')
    )
    return df

drivers = clean_column_names(drivers)
constructors = clean_column_names(constructors)
races = clean_column_names(races)
results = clean_column_names(results)
qualifying = clean_column_names(qualifying)
F1Drivers = clean_column_names(F1Drivers)
```

### 4.1 Handle Missing Values  
We apply simple, safe strategies:
- Drop duplicate rows  
- Fill missing text fields with `"Unknown"`  
- Leave complex missing values for later modelling  

```python
# Drop duplicates
drivers.drop_duplicates(inplace=True)
constructors.drop_duplicates(inplace=True)
races.drop_duplicates(inplace=True)
results.drop_duplicates(inplace=True)
qualifying.drop_duplicates(inplace=True)
F1Drivers.drop_duplicates(inplace=True)

# Fill simple missing values
drivers.fillna({'nationality': 'Unknown'}, inplace=True)
constructors.fillna({'constructorref': 'unknown'}, inplace=True)
```

### 4.2 Fix Data Types  
Convert date fields to datetime and ensure IDs are integers.

```python
# Convert dates
races['date'] = pd.to_datetime(races['date'], errors='coerce')

# Convert IDs to integers where possible
int_columns = ['driverid', 'constructorid', 'raceid']
for col in int_columns:
    for df in [drivers, constructors, races, results, qualifying]:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int64')
```

### 4.3 Normalise Key Fields  
Standardise driver names and constructor names for consistency.

```python
drivers['fullname'] = (
    drivers['forename'].str.strip() + ' ' + drivers['surname'].str.strip()
)

constructors['name'] = constructors['name'].str.strip().str.title()
```

## 5. Missing Value Check  
Assess each dataset for missing values to identify potential cleaning steps.

```python
# Check for missing values

print("Missing values:")
for df_name, df in [('drivers', drivers), ('constructors', constructors), ('races', races), ('results', results), ('qualifying', qualifying), ('F1Drivers', F1Drivers)]:
    print(f"{df_name}: {df.isnull().sum().sum()} missing values")
```

## 6. Duplicate Check  
Check for duplicate rows in each dataset to ensure data integrity.

```python
# Check for duplicates
print("Duplicate rows:")
for df_name, df in [('drivers', drivers), ('constructors', constructors), ('races', races), ('results', results), ('qualifying', qualifying), ('F1Drivers', F1Drivers)]:
    print(f"{df_name}: {df.duplicated().sum()} duplicates")
```

## 7. Key Field Review  
Compare driver and constructor IDs across datasets to identify mismatches before merging.

```python
# Check ID consistency across datasets
print("Unique driver IDs in results vs drivers:")
print(set(results['driverid']) - set(drivers['driverid']))

print("Unique constructor IDs in results vs constructors:")
print(set(results['constructorid']) - set(constructors['constructorid']))
```

## 8. Merge Datasets  
Now that the raw data has been cleaned and validated, we merge the key datasets into a single analytical table.  
This unified dataset will be used for EDA and modelling.

### 8.1 Merge `drivers` with `F1Drivers_Dataset`
Use the `driver` column from F1Drivers as the full name field.


```python
# Prepare name fields for merging
drivers['fullname'] = drivers['forename'] + ' ' + drivers['surname']
F1Drivers['fullname'] = F1Drivers['driver'].str.strip()

# Merge on fullname
drivers_merged = drivers.merge(
    F1Drivers.drop(columns=['driver']), 
    on='fullname', 
    how='left'
)

drivers_merged.head().style.hide(subset="seasons", axis=1)
```

### 8.2 Merge `results` with enriched driver data  
This adds driver attributes to each race result.

```python
results_drivers = results.merge(
    drivers_merged,
    on='driverid',
    how='left'
)

results_drivers.head().style.hide(subset="seasons", axis=1)
```

### 8.3 Merge constructor information  
Adds constructor/team details to each race result.

```python
results_drivers_constructors = results_drivers.merge(
    constructors,
    on='constructorid',
    how='left'
)

results_drivers_constructors.head().style.hide(subset="seasons", axis=1)
```

### 8.4 Merge race information  
Adds race metadata such as circuit, date, and round.

```python
results_full = results_drivers_constructors.merge(
    races,
    on='raceid',
    how='left'
)

results_full.head().style.hide(subset="seasons", axis=1)
```

### 8.5 Merge qualifying results  
Adds qualifying position and times to the unified dataset.

```python
results_full_main = results_full.merge(
    qualifying[['raceid', 'driverid', 'position', 'q1', 'q2', 'q3']],
    on=['raceid', 'driverid'],
    how='left'
)

results_full_main.head().style.hide(subset="seasons", axis=1)
```

### 8.6 Feature Engineering (Basic)  
We create two simple features:
- **driver_age** at the time of the race  
- **driver_experience** (number of races completed before this one)

```python
# Convert driver DOB to datetime
results_full['dob'] = pd.to_datetime(results_full['dob'], errors='coerce')

# Driver age at race
results_full['driver_age'] = (
    (results_full['date'] - results_full['dob']).dt.days / 365.25
)
```

```python
results_full = results_full.sort_values(['driverid', 'date'])

results_full['driver_experience'] = (
    results_full.groupby('driverid').cumcount()
)
```

### 8.7 Save Processed Dataset  
We export the unified dataset to `data/processed/` for use in EDA and modelling.

```python
output_path = '../data/processed/merged_f1_dataset.csv'
results_full.to_csv(output_path, index=False)

print("Saved processed dataset to:", output_path)
```

## 9. Light Cleanup Pass (Final Polish)

With all datasets successfully merged, this final cleanup step prepares the data for exploratory analysis and modelling.

This pass focuses on:
- assigning clear, meaningful column names
- standardising race and qualifying position fields
- removing unused or redundant columns
- organising the dataset into logical groups
- exporting a clean, modellingready CSV

This ensures that downstream notebooks (EDA, modelling, dashboarding) work with a tidy, wellstructured dataset.
 

## 9.1 Rename Columns

The merged dataset originally contained ambiguous or duplicated column names inherited from multiple source tables.
This step standardises the naming so each field has a clear purpose.
Key improvements include:
- converting raw fields (positionOrder, grid) into intuitive names (race_position, qualifying_position)
- replacing generic names like name_x and name_y with descriptive labels (constructor_name, circuit_name)
- aligning driver and constructor metadata under consistent naming conventions
- ensuring race timing fields (time_x, time_y) are clearly distinguished

This renaming step makes the dataset far easier to interpret and eliminates the need to remember specific naming quirks


```python
clean_df = results_full.copy()

# Rename columns for clarity
rename_map = {
    'positionorder': 'race_position',        # finishing position (numeric)
    'grid': 'qualifying_position',           # qualifying position (grid)
    'points': 'race_points',                 # from results table
    'name_x': 'constructor_name',
    'name_y': 'circuit_name',
    'nationality_x': 'driver_nationality',
    'nationality_y': 'constructor_nationality',
    'time_x': 'race_time',
    'time_y': 'race_start_time',
    'number_x': 'race_number',
    'number_y': 'driver_number'
}

clean_df.rename(columns=rename_map, inplace=True)
```

## 9.2 Drop Irrelevant Columns
These fields are not needed for EDA or modelling


```python
cols_to_drop = [
    'url_x', 'url_y', 'fp1_date', 'fp1_time', 'fp2_date', 'fp2_time',
    'fp3_date', 'fp3_time', 'quali_date', 'quali_time',
    'sprint_date', 'sprint_time'
]

clean_df.drop(columns=[c for c in cols_to_drop if c in clean_df.columns], inplace=True)
```

## 9.3 Reorder Columns
This groups the dataset into logical sections:
- race info
- driver info
- constructor info
- qualifying
- race results
- engineered features


```python
ordered_columns = [
    # Race metadata
    'raceid', 'year', 'round', 'circuitid', 'circuit_name', 'date', 'race_start_time',

    # Driver metadata
    'driverid', 'driverref', 'code', 'forename', 'surname', 'fullname',
    'dob', 'driver_age', 'driver_nationality',

    # Enriched driver stats
    'seasons', 'championships', 'race_entries', 'race_starts',
    'pole_positions', 'race_wins', 'podiums', 'fastest_laps',
    'career_points', 'years_active',

    # Constructor metadata
    'constructorid', 'constructorref', 'constructor_name', 'constructor_nationality',

    # Qualifying
    'qualifying_position',

    # Race results
    'race_number', 'grid', 'race_position', 'race_points', 'laps',
    'race_time', 'milliseconds', 'fastestlap', 'fastestlaptime',
    'fastestlapspeed', 'statusid',

    # Engineered features
    'driver_experience'
]

# Keep only columns that exist (avoids errors)
ordered_columns = [c for c in ordered_columns if c in clean_df.columns]

clean_df = clean_df[ordered_columns]
```
